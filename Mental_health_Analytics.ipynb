```python
# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(df.Sleep_Quality,df.Work_Life_Balance_Rating, marker='o', linestyle='-', color='b', linewidth=2, markersize=8)

# Add labels and title
plt.xlabel("Sleep Quality")
plt.ylabel("Average Work-Life Balance Rating")
plt.title("Work-Life Balance Rating by Sleep Quality")
plt.grid(True)

plt.show()
```

```python
# Create a violin plot
plt.figure(figsize=(10, 8))
sns.violinplot(x="Job_Role", y="Industry", data=df)

# Add title and labels
plt.title("Violin Plot of Job Role by Industry")
plt.xlabel("Job Role")
plt.ylabel("Industry")

# Show the plot
plt.show()
```

```python
# Create a histogram of the flipper length
sns.histplot(data=df, x='Gender', bins=80, kde=True)

# Add title and labels
plt.title('Histogram of Gender Lengths')
plt.xlabel('Gender Length (mm)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Create a bar plot of average fare by passenger class
plt.figure(figsize=(10, 10))
sns.barplot(data=df, x='Job_Role', y='Industry')

# Add title and labels
plt.title('Job Role By Industry')
plt.xlabel('Job ROle')
plt.ylabel('Industry')
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```
